```python
tickers = ['AAPL', "TWTR", "KO", "^GSPC"]
df = yf.download(tickers, period='1y', interval='1d')
df = df.T
df = df.loc['Adj Close']
df = df.T

data = {}
for ticker in tickers:
     data[ticker] =  ((1 + df[ticker].pct_change(1).dropna()).cumprod() - 1) * 100
```

```python
x = np.linspace(3, 13, 11)
```

```python
def normal(x):
     return np.sqrt(2 * x - 6)

def inverse(x):
     return  abs(0.5 * x ** 2 + 3)

def normalizeData(data):
    return (data - np.min(data)) / (np.max(data) - np.min(data))
```

```python
plt.figure(figsize=(12, 6), dpi=200)

plt.plot(x,normalizeData(np.vectorize(normal)(x)), label='f1')
plt.plot(x, np.vectorize(inverse)(x), label='f-1')
plt.plot(x, x, label='y=x')
plt.legend()
```
